```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
from math import sin

from hpgmg.finite_volume.simple_hpgmg import SimpleMultigridSolver
```

```python
def plot(surface, shape=None, title='surface', use_heat=True):
    surface = surface if not shape else surface.reshape(shape)
    xs, ys = np.indices(surface.shape)
    
    if not use_heat:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(xs, ys, surface)
        plt.title("{}, average is {}".format(title, np.average(surface)))
    else:
        fig = plt.figure(figsize=(6, 3.2))
        ax = fig.add_subplot(111)
        ax.set_title("{}, average is {:8.4}".format(title, np.average(surface)))
        plt.imshow(surface)  # , vmin=-30, vmax=30)
        ax.set_aspect('equal')

        cax = fig.add_axes([0.12, 0.1, 0.78, 0.8])
        cax.get_xaxis().set_visible(False)
        cax.get_yaxis().set_visible(False)
        cax.patch.set_alpha(0)
        cax.set_frame_on(False)
        plt.colorbar(orientation='vertical')
        plt.show()
```

```python
solver = SimpleMultigridSolver.get_solver(["7", "--dimensions", "2", "-nv", "1", "--log"])
```

```python
# print(solver.fine_level.cell_values.space)
# plot(solver.fine_level.exact_solution, title="initial heat")
print(solver.fine_level.mean_mesh(solver.fine_level.exact_solution))
surface = np.array(solver.fine_level.exact_solution)
plot(surface, title="exact solution")
surface = np.array(solver.fine_level.right_hand_side)
plot(surface, title="right hand side")
solver.solve()
surface = np.array(solver.fine_level.cell_values)
plot(surface, title="solved values")
surface = np.array(solver.fine_level.residual)
plot(surface, title="residual")
```
