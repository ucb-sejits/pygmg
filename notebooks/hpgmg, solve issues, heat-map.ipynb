```python
def plot(surface, shape=None, title='surface'):
    surface = surface if not shape else surface.reshape(shape)
    xs, ys = np.indices(surface.shape)
    
#     fig = plt.figure()
#     ax = fig.add_subplot(111, projection='3d')
#     ax.plot_surface(xs, ys, surface)
#     plt.title("{}, average is {}".format(title, np.average(surface)))
    
    fig = plt.figure(figsize=(6, 3.2))
    ax = fig.add_subplot(111)
    ax.set_title("{}, average is {}".format(title, np.average(surface)))
    plt.imshow(surface)
    ax.set_aspect('equal')

    cax = fig.add_axes([0.12, 0.1, 0.78, 0.8])
    cax.get_xaxis().set_visible(False)
    cax.get_yaxis().set_visible(False)
    cax.patch.set_alpha(0)
    cax.set_frame_on(False)
    plt.colorbar(orientation='vertical')
    plt.show()
```

```python
def iterate(solver, trials=10, plots = 3):
    x = np.zeros_like(b)
    plot_interval = trials / (plots+1) # we show last plot
    for trial in range(trials):
        new_x = solver(A, b, x)

        if trial % 200 == 0:
            residual = x - new_x.reshape(b.shape)
            # plot(residual, b.shape, "iteration {}".format(trial))
            plot(new_x, b.shape, "iteration {}".format(trial))
        x = new_x
    
    plot(x, b.shape, "iteration {}".format(trial))
        
```
