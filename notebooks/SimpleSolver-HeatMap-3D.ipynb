```python
from __future__ import print_function
%matplotlib inline
import numpy as np
from hpgmg.finite_volume.ipython_tools import plot

from math import sin

from hpgmg.finite_volume.simple_hpgmg import SimpleMultigridSolver
```

```python
solver = SimpleMultigridSolver.get_solver([
    "4", 
    "--dimensions", "3", 
    "--bottom-solver", "smoother",
    "--smoother", "j",
    "--equation", "p",
    "--boundary-condition", "p",
    "--number-of-vcycles", "2", 
    "--log"
])
level = solver.fine_level
```

```python
plot(level.exact_solution[1], title="exact solution")
plot(level.right_hand_side[1], title="right hand side")
solver.solve()
plot(level.cell_values[1], title="solved values")
plot(level.temp[1], title="residual")
```
