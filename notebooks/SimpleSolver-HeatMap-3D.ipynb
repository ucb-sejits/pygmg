```python
from __future__ import print_function
%matplotlib inline
import numpy as np
from hpgmg.finite_volume.ipython_tools import plot

from math import sin

from hpgmg.finite_volume.simple_hpgmg import SimpleMultigridSolver
```

```python
size=4
solver = SimpleMultigridSolver.get_solver([
    "{}".format(size), 
    "--dimensions", "3", 
    "--bottom-solver", "smoother",
    "--smoother", "j",
    "--equation", "h",
    "--boundary-condition", "p",
    "--number-of-vcycles", "10", 
    "--problem", "sine",
    "--log"
])
level = solver.fine_level
```

```python
plane_index = size / 2
plot(level.exact_solution[plane_index], title="exact solution")
plot(level.right_hand_side[plane_index], title="right hand side")
solver.solve()
plot(level.cell_values[plane_index], title="solved values")
solver.show_error_information()
```

```python
solver.show_timing_information()
solver.show_error_information()
solver.run_richardson_test()
```
