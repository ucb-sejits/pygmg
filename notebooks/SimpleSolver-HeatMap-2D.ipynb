```python
from __future__ import print_function
%matplotlib inline
import numpy as np
from hpgmg.finite_volume.ipython_tools import plot

from hpgmg.finite_volume.simple_hpgmg import SimpleMultigridSolver
from hpgmg.finite_volume.operators.jacobi_smoother import JacobiSmoother
```

```python
solver = SimpleMultigridSolver.get_solver([
    "6", 
    "--dimensions", "2", 
    "--smoother", "j",
    "--use-l1-jacobi", 
    "--number-of-vcycles", "6", 
    "--boundary-condition", "d",
    "--problem-name", "p4",
#     "--problem-name", "sine",
    "--variable-coefficient",
    "--log",
])
level = solver.fine_level
```

```python
plot(level.exact_solution, title="exact solution")
plot(level.right_hand_side, title="right hand side")
if solver.is_variable_coefficient:
    plot(level.beta_face_values[0], title="beta_face_i")
    plot(level.beta_face_values[1], title="beta_face_j")
solver.problem_operator.rebuild_operator(level)
if isinstance(solver.smoother, JacobiSmoother):
    if solver.smoother.use_l1_jacobi:
        print("Using Jacobi smoother with l1_inverse")
        plot(level.l1_inverse, title="l1_inverse")
    else:
        print("Using Jacobi smoother with d_inverse")
        plot(level.d_inverse, title="d_inverse")
else:
    print("Using Chebyshev smoother with d_inverse")
    plot(level.d_inverse, title="d_inverse")
    
# plot(level.valid, title="valid", ghost_size=0)
```

```python
solver.solve()
plot(level.cell_values, title="solved values")
plot(level.residual, title="residual")
```
